```python
#input数据：itemId（List,50条数据); title（List,50条数据）；pics(List,50条数据); author(List,50条数据); category(List,50条数据)
#output：result字典，里面包含了itemId，title，pics，author，category互相对应的前10条数据


#feeditem
# 思路，按照给定的inputitem，先提取第一条数据。

# 查找符合2.2要求，相同author不相邻原则，查找第二条数据。e.g：如果inputitem第一条数据和第二条数据作者相同，那我的output数据应该是
# skip第二条数据，输出第一条和第三条数据，前提是第三条数据的author和第一条数据的author不一样  input：【1，1，2，3】 output：【1，2，3】 

# 查找符合2.3要求，相同category的输出item不能连续超过2条。查找第三条数据。e.g.假如input：【1，1，1，2，3】 output：【1，1，2，3】
# 如果查找到了符合2.3要求的数据，但是和2.2冲突时，优先考虑2.2

# 如果到最后都没选到10条符合要求的数据，则尽可能多的筛选出符合要求的前十条数据，在追加不符合要求的，直到补充满10条为止。

#初始化数据
itemId,title,pics,author,category = [],[],[],[],[]
for i in range(50):
    itemId.append(i+1)
    title.append(i+100)
    pics.append(i+1000)
    
for i in range(51,101):
    author.append(i)
    #更改1，2位置和0位置为同一作者，第5位置和第4位置为同一作者
author[1] = author[0]
author[2] = author[0]
author[5] = author[4]

# for i in range(50): # 全部同一个作者
#     author.append(2)

for i in range(5):
    for j in range(10):
        category.append(j+1)
category[1] = category[0]
category[2] = category[0]
category[3] = category[0]
category[4] = category[0]
category[8] = category[7]
category[9] = category[7]

# for i in range(50): #全部同一种类
#     category.append(1)

#函数实现
def feedItem(itemId,title,pics,author,category):
    #result 初始值输入第一条数据
    count = 0
    index =  [0]
    result = {'itemId':[itemId[0]],
             'title':[title[0]],
             'pics':[pics[0]],
             'author':[author[0]],
             'category':[category[0]]}
    for i in range(1,len(itemId)):
        #第二条数据判定作者是否和第一条数据相同，不相同的话就输入第二条数据
        if len(result['itemId']) == 1:
            if author[i] != author[i-1]:
                index.append(i)
                result['itemId'] += [itemId[i]]
                result['title'] += [title[i]]
                result['pics'] += [pics[i]]
                result['author'] += [author[i]]
                result['category'] += [category[i]]
        else:
            # 假若一直循环到最后都是跳过，则忽略category选项，输入符合相同作者不相临的数据
            if i == len(itemId)-1:
                #print('-----count: ',count)
                for j in range(len(itemId)-count-2,len(itemId)):
                    if len(result['itemId']) >= 10:
                        return result
                    if author[j] != author[j-1]:
                        index.append(j)
                        result['itemId'] += [itemId[j]]
                        result['title'] += [title[j]]
                        result['pics'] += [pics[j]]
                        result['author'] += [author[j]]
                        result['category'] += [category[j]]
            else:
                # 第三条数据 判定作者是否和前面的相同，不相同的情况下追加判定：
                if author[i] != author[i-1]:
                    #       判断当前分类是否和上一个一样，若不一样，第三条数据则可直接输入。
                    if category[i] != category[i-1]:
                        index.append(i)
                        result['itemId'] += [itemId[i]]
                        result['title'] += [title[i]]
                        result['pics'] += [pics[i]]
                        result['author'] += [author[i]]
                        result['category'] += [category[i]]
                    #   若一样，则判断当前分类和上上个分类是否一样，若一样则跳过当前数据，若不一样则输入数据。
                    elif category[i] == category[i-1] and category[i-1] != category[i-2]:
                        index.append(i)
                        result['itemId'] += [itemId[i]]
                        result['title'] += [title[i]]
                        result['pics'] += [pics[i]]
                        result['author'] += [author[i]]
                        result['category'] += [category[i]]
                    else:
                        count += 1
                        continue
        # 得到10组数据后跳出循环
        if len(result['itemId']) >= 10:
            return result

    #假若忽略category选项后还不能补全10条消息，则按照input顺序补剩下的，直至到10条为止
    for k in range(len(itemId)):
        if len(result['itemId']) < 10:
            if k not in index:
                result['itemId'] += [itemId[k]]
                result['title'] += [title[k]]
                result['pics'] += [pics[k]]
                result['author'] += [author[k]]
                result['category'] += [category[k]]
        else:
            return result
feedItem(itemId,title,pics,author,category)
```
